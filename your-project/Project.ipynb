# Testing if the Volume of Avocado sold per type are uniformly distributed:

```python
observed = np.array(df.groupby("type")["total volume"].sum())
observed
```

```python
# expected frequencies --> assuming that quantity sold is uniformly distributed

expected = np.array([observed.sum()/2, observed.sum()/2])
expected
```

```python
# perform the chi-squared test

stats.chisquare(observed, expected)

# tells us that there is a strong evidence to reject the null hypothesis
# --> the quantity of avocados sold is not uniformly distributed of the type
```

# Hypothesis test:
Sales of bags > Sales of invidivual ---> so our H0: Aver ind > Aver bgas

```python
#create a new column with the individual total sales
df["total individual"] = df["total volume"] - df["total bags"]
```

```python
#Calculate the average of the total individual avocados sold
np.mean(df["total individual"])
```

```python
#Calculate the average of the total bags avocados sold
np.mean(df["total bags"])
```

```python
#Perform the t test

#doc: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

from scipy.stats import ttest_ind

ttest_ind(df["total bags"], df["total individual"], alternative = "greater")
```

```python
#Conclusion:

if 1<0.05:
    print("Reject null hypothesis")
    print("Reject that the mean of the individual avocados sold is lower than the mean of the bags of avocados sold.")
else:
    print("We cannot reject the null hypothesis")
    print("That that the mean of the individual avocados sold is higher than the average of the bags of avocados sold.")
```
